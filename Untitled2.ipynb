```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler
import math

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Generate synthetic time series data with combined pattern
def generate_synthetic_data(n_samples=1000):
    # Time steps
    t = np.linspace(0, 10, n_samples)
    
    # Combined pattern: 0.5*sin(x) + exp(x/10) + noise
    combined_pattern = 0.5 * np.sin(t) + np.exp(t/10)
    noise = np.random.normal(0, 0.1, n_samples)
    combined_series = combined_pattern + noise
    
    # Create DataFrame
    df = pd.DataFrame({
        'time': t,
        'value': combined_series
    })
    
    return df

# Generate test data with similar pattern but slight variations
def generate_test_data(n_samples=200):
    # Time steps (continuing from where training data ends)
    t = np.linspace(10, 12, n_samples)
    
    # Similar pattern with slight variations: 0.45*sin(x) + exp(x/10) + noise
    # Small change in sine amplitude and slightly different noise profile
    combined_pattern = 0.45 * np.sin(t) + np.exp(t/10)
    noise = np.random.normal(0, 0.12, n_samples)  # Slight increase in noise
    combined_series = combined_pattern + noise
    
    # Create DataFrame
    df = pd.DataFrame({
        'time': t,
        'value': combined_series
    })
    
    return df

# Create sequences for time series prediction
def create_sequences(data, seq_length, pred_length=1):
    X, y = [], []
    for i in range(len(data) - seq_length - pred_length + 1):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length:i+seq_length+pred_length])
    return np.array(X), np.array(y)

# Prepare data for PyTorch model
def prepare_data_for_model(train_series, test_series, seq_length, pred_length=1, batch_size=32):
    # Scale the data using training data statistics
    scaler = MinMaxScaler(feature_range=(0, 1))
    train_scaled = scaler.fit_transform(train_series.values.reshape(-1, 1))
    test_scaled = scaler.transform(test_series.values.reshape(-1, 1))
    
    # Create sequences
    X_train, y_train = create_sequences(train_scaled, seq_length, pred_length)
    X_test, y_test = create_sequences(test_scaled, seq_length, pred_length)
    
    # Reshape y to handle the dimension issue
    if y_train.shape[2] == 1:
        y_train = y_train.reshape(y_train.shape[0], y_train.shape[1])
    if y_test.shape[2] == 1:
        y_test = y_test.reshape(y_test.shape[0], y_test.shape[1])
    
    # Convert to PyTorch tensors
    X_train_tensor = torch.FloatTensor(X_train)
    y_train_tensor = torch.FloatTensor(y_train)
    X_test_tensor = torch.FloatTensor(X_test)
    y_test_tensor = torch.FloatTensor(y_test)
    
    # Create DataLoader
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)
    
    return train_loader, test_loader, scaler

# Define LSTM model for time series prediction
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=1, output_size=1):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.output_size = output_size
        
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        # Initialize hidden state and cell state
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        
        # Forward propagate LSTM
        out, _ = self.lstm(x, (h0, c0))
        
        # Get output from the last time step
        out = self.fc(out[:, -1, :])
        return out

# Training function
def train_model(model, train_loader, test_loader, epochs=50):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    train_losses = []
    test_losses = []
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            
            # Forward pass
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            
            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            
        # Calculate average training loss
        train_loss = train_loss / len(train_loader)
        train_losses.append(train_loss)
        
        # Evaluate on test set
        model.eval()
        test_loss = 0
        with torch.no_grad():
            for X_batch, y_batch in test_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                outputs = model(X_batch)
                loss = criterion(outputs, y_batch)
                test_loss += loss.item()
                
        test_loss = test_loss / len(test_loader)
        test_losses.append(test_loss)
        
        if (epoch+1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}')
    
    return model, train_losses, test_losses

# Make predictions with the trained model
def make_predictions(model, data_loader, scaler):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    
    predictions = []
    actual = []
    
    with torch.no_grad():
        for X_batch, y_batch in data_loader:
            X_batch = X_batch.to(device)
            outputs = model(X_batch)
            
            # Move tensors to CPU if needed
            outputs = outputs.cpu().numpy()
            y_batch = y_batch.cpu().numpy()
            
            predictions.extend(outputs)
            actual.extend(y_batch)
    
    # Reshape predictions and actual values for inverse transform if needed
    predictions_np = np.array(predictions)
    actual_np = np.array(actual)
    
    # Check if reshaping is needed for inverse transform
    if len(predictions_np.shape) == 2 and predictions_np.shape[1] > 1:
        # Need to reshape for multi-step predictions to apply inverse transform
        reshaped_preds = []
        for pred in predictions_np:
            # Reshape each prediction to match scaler's expected format
            reshaped_preds.extend(pred.reshape(-1, 1))
        predictions_inv = scaler.inverse_transform(np.array(reshaped_preds))
        
        # Reshape back to original structure
        predictions_inv = predictions_inv.reshape(predictions_np.shape)
    else:
        # For single step prediction or already correct shape
        predictions_inv = scaler.inverse_transform(predictions_np.reshape(-1, 1))
    
    # Same process for actual values
    if len(actual_np.shape) == 2 and actual_np.shape[1] > 1:
        reshaped_actual = []
        for act in actual_np:
            reshaped_actual.extend(act.reshape(-1, 1))
        actual_inv = scaler.inverse_transform(np.array(reshaped_actual))
        actual_inv = actual_inv.reshape(actual_np.shape)
    else:
        actual_inv = scaler.inverse_transform(actual_np.reshape(-1, 1))
    
    return predictions_inv, actual_inv

# Main execution
if __name__ == "__main__":
    # Generate the synthetic data
    train_data = generate_synthetic_data(1000)
    test_data = generate_test_data(200)
    
    # Visualize the data
    plt.figure(figsize=(12, 6))
    plt.plot(train_data['time'], train_data['value'], label='Training Data')
    plt.plot(test_data['time'], test_data['value'], label='Test Data')
    plt.title('Combined Pattern: 0.5*sin(x) + exp(x/10) + noise')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Parameters
    seq_length = 20
    pred_length = 5
    batch_size = 32
    epochs = 50
    
    # Prepare data for model
    train_loader, test_loader, scaler = prepare_data_for_model(
        train_data['value'], test_data['value'], seq_length, pred_length, batch_size
    )
    
    # Define and train model
    model = LSTMModel(input_size=1, hidden_size=64, num_layers=2, output_size=pred_length)
    model, train_losses, test_losses = train_model(model, train_loader, test_loader, epochs)
    
    # Make predictions
    predictions, actual = make_predictions(model, test_loader, scaler)
    
    # Calculate RMSE
    if len(predictions.shape) > 1 and predictions.shape[1] > 1:
        rmse = math.sqrt(np.mean((predictions[:, 0] - actual[:, 0]) ** 2))
    else:
        rmse = math.sqrt(np.mean((predictions - actual) ** 2))
    print(f'RMSE: {rmse:.4f}')
    
    # Plot results
    plt.figure(figsize=(12, 6))
    if len(predictions.shape) > 1 and predictions.shape[1] > 1:
        plt.plot(actual[:100, 0], label='Actual', marker='o', markersize=3)
        plt.plot(predictions[:100, 0], label='Predicted', alpha=0.7, marker='x', markersize=3)
    else:
        plt.plot(actual[:100], label='Actual', marker='o', markersize=3)
        plt.plot(predictions[:100], label='Predicted', alpha=0.7, marker='x', markersize=3)
    plt.title('Time Series Prediction Results')
    plt.xlabel('Time Step')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Plot training losses
    plt.figure(figsize=(12, 6))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(test_losses, label='Test Loss')
    plt.title('Model Loss During Training')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Plot multiple-step prediction example
    if pred_length > 1:
        plt.figure(figsize=(12, 6))
        # Take a single example and show all predicted steps
        example_idx = 10
        plt.plot(range(pred_length), actual[example_idx], 'b-o', label='Actual')
        plt.plot(range(pred_length), predictions[example_idx], 'r-x', label='Predicted')
        plt.title(f'Multi-step Prediction Example (Sample {example_idx})')
        plt.xlabel('Time Step')
        plt.ylabel('Value')
        plt.legend()
        plt.grid(True)
        plt.show()
```
