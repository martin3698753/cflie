```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
def generate_synthetic_data(seq_length, num_samples, noise_level=0.1):
    x = np.linspace(0, 100, num_samples)
    y = np.sin(x) + noise_level * np.random.randn(num_samples)
    return y

# Prepare data function
def create_sequences(data, seq_length, forecast_factor):
    X, Y = [], []
    out_length = seq_length * forecast_factor
    for i in range(len(data) - seq_length - out_length):
        X.append(data[i:i+seq_length])
        Y.append(data[i+seq_length:i+seq_length+out_length])
    X = np.array(X)
    Y = np.array(Y)
    return torch.tensor(X, dtype=torch.float32).unsqueeze(1), torch.tensor(Y, dtype=torch.float32)

# Define Model
class ConvLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(ConvLSTM, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.lstm = nn.LSTM(16, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        x = self.conv1(x)
        x = torch.relu(x)
        x = x.permute(0, 2, 1)  # (batch, seq, channels) for LSTM
        lstm_out, _ = self.lstm(x)
        out = self.fc(lstm_out[:, -1, :])
        return out

# Hyperparameters
seq_length = 20
a = 3
forecast_length = seq_length * a
hidden_size = 50
num_layers = 2
learning_rate = 0.001
num_epochs = 50

# Generate and prepare data
data = generate_synthetic_data(200, 1000)
X, Y = create_sequences(data, seq_length, a)
train_size = int(0.8 * len(X))
X_train, Y_train = X[:train_size], Y[:train_size]
X_test, Y_test = X[train_size:], Y[train_size:]

# Model initialization
model = ConvLSTM(seq_length, hidden_size, num_layers, forecast_length)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    optimizer.zero_grad()
    output = model(X_train)
    loss = criterion(output, Y_train)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}')

# Prediction
model.eval()
predictions = model(X_test).detach().numpy()

# Visualization
plt.figure(figsize=(12, 5))
plt.plot(range(len(data)), data, label='True Data', alpha=0.5)
plt.plot(range(train_size + seq_length, train_size + seq_length + len(predictions)), predictions[:, 0], label='Predictions', linestyle='dashed')
plt.legend()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Generate more complex synthetic data
def generate_synthetic_data(num_samples, noise_level=0.2, test=False):
    x = np.linspace(0, 100, num_samples)
    y = np.sin(x) + 0.3 * np.cos(2 * x) + 0.05 * np.exp(0.02 * x) + noise_level * np.random.randn(num_samples)
    if test:
        y += 0.5 * np.sin(x)  # Make test data slightly different but similar
    return y

# Prepare data function
def create_sequences(data, seq_length, forecast_factor):
    X, Y = [], []
    out_length = seq_length * forecast_factor
    for i in range(len(data) - seq_length - out_length):
        X.append(data[i:i+seq_length])
        Y.append(data[i+seq_length:i+seq_length+out_length])
    X = np.array(X)
    Y = np.array(Y)
    return torch.tensor(X, dtype=torch.float32).unsqueeze(1), torch.tensor(Y, dtype=torch.float32)

# Define Model with additional layers and dropout for accuracy improvement
class ConvLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(ConvLSTM, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=32, kernel_size=5, padding=2)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=16, kernel_size=3, padding=1)
        self.lstm = nn.LSTM(16, hidden_size, num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        x = self.conv1(x)
        x = torch.relu(x)
        x = self.conv2(x)
        x = torch.relu(x)
        x = x.permute(0, 2, 1)  # (batch, seq, channels) for LSTM
        lstm_out, _ = self.lstm(x)
        out = self.fc(lstm_out[:, -1, :])
        return out

# Hyperparameters
seq_length = 30
a = 3
forecast_length = seq_length * a
hidden_size = 100
num_layers = 3
learning_rate = 0.0005
num_epochs = 100

# Generate train and test data
train_data = generate_synthetic_data(1200, noise_level=0.3)
test_data = generate_synthetic_data(1200, noise_level=0.3, test=True)

# Prepare sequences
X_train, Y_train = create_sequences(train_data, seq_length, a)
X_test, Y_test = create_sequences(test_data, seq_length, a)

# Model initialization
model = ConvLSTM(seq_length, hidden_size, num_layers, forecast_length)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    optimizer.zero_grad()
    output = model(X_train)
    loss = criterion(output, Y_train)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}')

# Evaluation
model.eval()
pred_train = model(X_train).detach().numpy()
pred_test = model(X_test).detach().numpy()

# Visualization
plt.figure(figsize=(12, 5))
plt.plot(range(len(train_data)), train_data, label='Train Data', alpha=0.5)
plt.plot(range(seq_length, seq_length + len(pred_train)), pred_train[:, 0], label='Train Predictions', linestyle='dashed')
plt.legend()
plt.title("Train Data vs Predictions")
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(range(len(test_data)), test_data, label='Test Data', alpha=0.5)
plt.plot(range(seq_length, seq_length + len(pred_test)), pred_test[:, 0], label='Test Predictions', linestyle='dashed')
plt.legend()
plt.title("Test Data vs Predictions")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

def norm(signal):
    norm_upper = 4
    norm_lower = 2
    signal = (signal - norm_lower) / (norm_upper - norm_lower)
    signal = signal[50:]
    return signal

def load_data(path_dir='data/5-2-25/'):
    t, signal = mt.battery(path_dir)
    return norm(signal)

class ResidualConv(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3):
        super().__init__()
        self.conv1 = nn.Conv1d(in_channels, out_channels, kernel_size, padding=1)
        self.bn1 = nn.BatchNorm1d(out_channels)
        self.conv2 = nn.Conv1d(out_channels, out_channels, kernel_size, padding=1)
        self.bn2 = nn.BatchNorm1d(out_channels)
        self.residual = nn.Conv1d(in_channels, out_channels, kernel_size=1) if in_channels != out_channels else nn.Identity()
    
    def forward(self, x):
        res = self.residual(x)
        x = torch.relu(self.bn1(self.conv1(x)))
        x = self.bn2(self.conv2(x))
        return torch.relu(x + res)

class ForecastingModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=2):
        super().__init__()
        self.conv = ResidualConv(1, 16)
        self.lstm = nn.LSTM(16, hidden_dim, num_layers, batch_first=True, bidirectional=True, dropout=0.2)
        self.attention = nn.Linear(hidden_dim * 2, 1)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
    
    def forward(self, x):
        x = self.conv(x)
        x = x.permute(0, 2, 1)  # (batch, seq, channels)
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = (lstm_out * attn_weights).sum(dim=1)
        return self.fc(context)

# Load Data
train_data = load_data()[2000:3000]
test_data = load_data()[1000:2000]
seq_length = 30
predict_length = seq_length * 2

def create_sequences(data, seq_length, predict_length):
    X, Y = [], []
    for i in range(len(data) - seq_length - predict_length):
        X.append(data[i:i + seq_length])
        Y.append(data[i + seq_length: i + seq_length + predict_length])
    return torch.tensor(np.array(X), dtype=torch.float32).unsqueeze(1), torch.tensor(np.array(Y), dtype=torch.float32)

X_train, Y_train = create_sequences(train_data, seq_length, predict_length)
X_test, Y_test = create_sequences(test_data, seq_length, predict_length)

# Model Training
model = ForecastingModel(input_dim=seq_length, hidden_dim=64, output_dim=predict_length)
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
criterion = nn.MSELoss()

epochs = 200
for epoch in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    loss = criterion(output, Y_train)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")

# Evaluation
with torch.no_grad():
    train_preds = model(X_train).numpy()
    test_preds = model(X_test).numpy()

# Visualization
plt.figure(figsize=(10, 4))
plt.plot(train_data, label='Train Data', alpha=0.6)
plt.plot(range(seq_length, seq_length + len(train_preds)), train_preds[:, 0], '--', label='Train Predictions')
plt.legend()
plt.title("Train Data vs Predictions")
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(test_data, label='Test Data', alpha=0.6)
plt.plot(range(seq_length, seq_length + len(test_preds)), test_preds[:, 0], '--', label='Test Predictions')
plt.legend()
plt.title("Test Data vs Predictions")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

# Dataset
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_length, a=1):
        self.data = torch.FloatTensor(data)
        self.seq_length = seq_length
        self.a = a

    def __len__(self):
        return len(self.data) - self.seq_length - self.a + 1

    def __getitem__(self, idx):
        x = self.data[idx:idx + self.seq_length]
        y = self.data[idx + self.seq_length + self.a - 1]
        return x, y

# Model
class TimeSeriesModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=256, num_layers=3, dropout=0.2):
        super(TimeSeriesModel, self).__init__()
        self.rnn = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, 
                          batch_first=True, dropout=dropout if num_layers > 1 else 0)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = x.unsqueeze(-1)  # (batch_size, seq_length, 1)
        x, _ = self.rnn(x)   # (batch_size, seq_length, hidden_size)
        x = x[:, -1, :]      # (batch_size, hidden_size)
        x = self.fc(x)       # (batch_size, 1)
        return x

# Prediction function for full series
def predict_full_series(model, data, seq_length):
    model.eval()
    predictions = []
    input_seq = torch.FloatTensor(data[:seq_length]).unsqueeze(0)  # (1, seq_length)
    
    with torch.no_grad():
        for i in range(len(data) - seq_length):
            pred = model(input_seq)  # (1, 1)
            predictions.append(pred.item())
            pred = pred.view(1, 1)  # Reshape to (1, 1)
            input_seq = torch.cat((input_seq[:, 1:], pred), dim=1)
    
    return predictions

# Normalize data
def normalize(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std, mean, std

# Denormalize data
def denormalize(data, mean, std):
    return data * std + mean

# Training and Visualization
def train_and_visualize():
    # Generate training data
    np.random.seed(42)
    t = np.linspace(0, 10, 1000)
    train_data = np.sin(t) + np.random.normal(0, 0.1, 1000)
    train_data, train_mean, train_std = normalize(train_data)

    # Generate test data (same base, different noise)
    np.random.seed(123)
    test_data = np.sin(t) + np.random.normal(0, 0.1, 1000)
    test_data = (test_data - train_mean) / train_std  # Normalize using train stats

    # Hyperparameters
    seq_length = 100  # Increased
    a = 1
    batch_size = 32
    epochs = 200  # Increased
    learning_rate = 0.001
    teacher_forcing_ratio = 0.5  # 50% chance to use predicted values during training

    # Training Dataset and DataLoader
    train_dataset = TimeSeriesDataset(train_data, seq_length=seq_length, a=a)
    train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    # Model, Loss, Optimizer, Scheduler
    model = TimeSeriesModel(input_size=1, hidden_size=256, num_layers=3, dropout=0.2)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=50, gamma=0.5)  # Reduce LR every 50 epochs

    # Track loss
    losses = []

    # Training loop with teacher forcing
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for x_batch, y_batch in train_dataloader:
            optimizer.zero_grad()
            outputs = model(x_batch)  # Initial prediction

            # Teacher forcing: sometimes use the model's prediction as input
            if torch.rand(1).item() < teacher_forcing_ratio:
                # Use the model's prediction for the next step
                x_next = x_batch.clone()
                x_next[:, -1] = outputs.squeeze()  # Replace last value with prediction
                outputs = model(x_next)  # Predict again with modified input

            loss = criterion(outputs.squeeze(), y_batch)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        scheduler.step()  # Update learning rate
        avg_loss = running_loss / len(train_dataloader)
        losses.append(avg_loss)
        print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}, LR: {scheduler.get_last_lr()[0]:.6f}")

    # Plot 1: Training Loss
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, epochs + 1), losses, label='Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Loss Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot 2: Train Data Full Series
    train_predictions = predict_full_series(model, train_data, seq_length)
    train_predictions = denormalize(np.array(train_predictions), train_mean, train_std)
    train_data = denormalize(train_data, train_mean, train_std)

    plt.figure(figsize=(12, 6))
    plt.plot(range(len(train_data)), train_data, label='Actual Train Series', alpha=0.5)
    plt.plot(range(seq_length, len(train_data)), train_predictions, 
             label='Predicted Train Series', color='red', linestyle='--')
    plt.xlabel('Time Step')
    plt.ylabel('Value')
    plt.title('Full Train Series Prediction')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot 3: Test Data Full Series
    test_predictions = predict_full_series(model, test_data, seq_length)
    test_predictions = denormalize(np.array(test_predictions), train_mean, train_std)
    test_data = denormalize(test_data, train_mean, train_std)

    plt.figure(figsize=(12, 6))
    plt.plot(range(len(test_data)), test_data, label='Actual Test Series', alpha=0.5)
    plt.plot(range(seq_length, len(test_data)), test_predictions, 
             label='Predicted Test Series', color='red', linestyle='--')
    plt.xlabel('Time Step')
    plt.ylabel('Value')
    plt.title('Full Test Series Prediction')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run
if __name__ == "__main__":
    train_and_visualize()
```
