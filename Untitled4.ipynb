```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
import random
from IPython.display import display, Math

# Generate artificial function mapping data
def g_function(x):
    return np.sin(x) + np.exp(0.1 * x) + np.random.normal(0, 0.1, x.shape)

def f_function(x):
    return 0.5 * x + 0.05 * x**2 + np.random.normal(0, 0.1, x.shape)

def generate_function_data(num_samples=1000, seq_length=20):
    X, Y = [], []
    for _ in range(num_samples):
        x_vals = np.linspace(0, 10, seq_length)
        X.append(g_function(x_vals))
        Y.append(f_function(x_vals))
    return np.array(X), np.array(Y)

# Custom Dataset class
class FunctionMappingDataset(Dataset):
    def __init__(self, X, Y):
        self.X = torch.tensor(X, dtype=torch.float32).unsqueeze(-1)
        self.Y = torch.tensor(Y, dtype=torch.float32).unsqueeze(-1)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]

# Define Encoder class
class Encoder(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
    
    def forward(self, x):
        _, (hidden, cell) = self.lstm(x)
        return hidden, cell

# Define Decoder class
class Decoder(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
    
    def forward(self, x, hidden, cell):
        output, (hidden, cell) = self.lstm(x, (hidden, cell))
        prediction = self.fc(output)
        return prediction, hidden, cell

# Display mathematical equations
display(Math(r'X = g(x) \in \mathbb{R}^{B \times T \times 1}'))
display(Math(r'Y = f(x) \in \mathbb{R}^{B \times T \times 1}'))
display(Math(r'H, C = \text{LSTM}_{enc}(X)'))
display(Math(r'Y, H, C = \text{LSTM}_{dec}(H, C)'))

# Define Seq2Seq Model
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, seq_length=20):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.seq_length = seq_length
    
    def forward(self, src, trg=None, teacher_forcing_ratio=0.5):
        batch_size = src.shape[0]
        hidden, cell = self.encoder(src)
        decoder_input = torch.zeros((batch_size, 1, 1), device=src.device)  # Start with zero input
        outputs = []
        
        for t in range(self.seq_length):
            decoder_output, hidden, cell = self.decoder(decoder_input, hidden, cell)
            outputs.append(decoder_output)
            decoder_input = decoder_output if trg is None or random.random() > teacher_forcing_ratio else trg[:, t:t+1, :]
        
        return torch.cat(outputs, dim=1)

# Create dataset
X, Y = generate_function_data()
dataset = FunctionMappingDataset(X, Y)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Initialize model
encoder = Encoder()
decoder = Decoder()
model = Seq2Seq(encoder, decoder)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# Training loop
num_epochs = 20
for epoch in range(num_epochs):
    total_loss = 0
    for src, trg in dataloader:
        optimizer.zero_grad()
        output = model(src, trg)
        loss = criterion(output, trg)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss / len(dataloader):.4f}")

# Visualization
model.eval()
sample_x, sample_y = dataset[0]
sample_x = sample_x.unsqueeze(0)  # Add batch dimension

with torch.no_grad():
    pred_y = model(sample_x).squeeze().numpy()

plt.plot(range(len(sample_x.squeeze())), sample_x.squeeze(), label="g(x) Input")
plt.plot(range(len(sample_x.squeeze())), sample_y.squeeze(), label="True f(x) Output")
plt.plot(range(len(sample_x.squeeze())), pred_y, label="Predicted f(x) Output", linestyle='dashed')
plt.legend()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
import random

def g_function(x):
    return np.sin(x) + np.exp(0.1 * x) + np.random.normal(0, 0.1, x.shape)

def f_function(x):
    return 0.5 * x + 0.05 * x**2 + np.random.normal(0, 0.1, x.shape)

class ChunkedFunctionDataset(Dataset):
    def __init__(self, num_samples=200, full_seq_length=100, chunk_size=20, overlap=0.5):
        self.chunk_size = chunk_size
        self.step_size = int(chunk_size * (1 - overlap))
        self.step_size = max(1, self.step_size)  # Ensure at least 1 step
        
        # Generate full sequences
        self.full_sequences = []
        for _ in range(num_samples):
            x_vals = np.linspace(0, 10, full_seq_length)
            g_vals = g_function(x_vals)
            f_vals = f_function(x_vals)
            self.full_sequences.append((g_vals, f_vals))
        
        # Precompute number of chunks per sequence
        self.chunks_per_sequence = (full_seq_length - chunk_size) // self.step_size + 1
        self.total_chunks = num_samples * self.chunks_per_sequence

    def __len__(self):
        return self.total_chunks

    def __getitem__(self, idx):
        # Determine which full sequence and which chunk within it
        seq_idx = idx // self.chunks_per_sequence
        chunk_idx = idx % self.chunks_per_sequence
        
        # Get the corresponding chunk
        start = chunk_idx * self.step_size
        end = start + self.chunk_size
        g_chunk, f_chunk = self.full_sequences[seq_idx]
        
        return (
            torch.tensor(g_chunk[start:end], dtype=torch.float32).unsqueeze(-1),
            torch.tensor(f_chunk[start:end], dtype=torch.float32).unsqueeze(-1)
        )

    def get_full_sequence(self, idx):
        """Get full unchunked sequence for evaluation"""
        g_vals, f_vals = self.full_sequences[idx]
        return (
            torch.tensor(g_vals, dtype=torch.float32).unsqueeze(-1),
            torch.tensor(f_vals, dtype=torch.float32).unsqueeze(-1)
        )

# Model Architecture
class Encoder(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
    
    def forward(self, x):
        _, (hidden, cell) = self.lstm(x)
        return hidden, cell

class Decoder(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
    
    def forward(self, x, hidden, cell):
        output, (hidden, cell) = self.lstm(x, (hidden, cell))
        prediction = self.fc(output)
        return prediction, hidden, cell

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
    
    def forward(self, src, trg=None, teacher_forcing_ratio=0.5):
        batch_size = src.shape[0]
        seq_len = src.shape[1]
        
        hidden, cell = self.encoder(src)
        decoder_input = torch.zeros((batch_size, 1, 1), device=src.device)
        outputs = []
        
        for t in range(seq_len):
            decoder_output, hidden, cell = self.decoder(decoder_input, hidden, cell)
            outputs.append(decoder_output)
            decoder_input = decoder_output if trg is None or random.random() > teacher_forcing_ratio else trg[:, t:t+1, :]
        
        return torch.cat(outputs, dim=1)

# Create dataset and dataloader
dataset = ChunkedFunctionDataset(
    num_samples=200,
    full_seq_length=100,
    chunk_size=20,
    overlap=0.5
)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Initialize model
encoder = Encoder()
decoder = Decoder()
model = Seq2Seq(encoder, decoder)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# Training loop
num_epochs = 20
for epoch in range(num_epochs):
    total_loss = 0
    for src, trg in dataloader:
        optimizer.zero_grad()
        output = model(src, trg)
        loss = criterion(output, trg)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss / len(dataloader):.4f}")
```

```python
def test_partial_sequence(model, dataset, partial_length=5):
    """Predict f(x) for the FIRST `partial_length` steps of g(x)."""
    model.eval()
    
    # Get sample and add batch dimension
    sample_x, sample_y = dataset[0]  # sample_x: [seq_len, 1], sample_y: [seq_len, 1]
    sample_x = sample_x.unsqueeze(0)  # [1, seq_len, 1]
    sample_y = sample_y.unsqueeze(0)  # [1, seq_len, 1]

    # Take partial input (first `partial_length` steps)
    partial_x = sample_x[:, :partial_length, :]  # [1, partial_length, 1]
    
    with torch.no_grad():
        # 1. Encode partial g(x)
        hidden, cell = model.encoder(partial_x)
        
        # 2. Initialize decoder
        decoder_input = torch.zeros((1, 1, 1), device=sample_x.device)
        outputs = []
        
        # 3. Decode AUTOREGRESSIVELY for the SAME LENGTH as input
        for t in range(partial_length):
            decoder_output, hidden, cell = model.decoder(decoder_input, hidden, cell)
            outputs.append(decoder_output)
            decoder_input = decoder_output  # Feed prediction back
        
        # Combine predictions: [partial_length, 1] -> [partial_length]
        pred_y = torch.cat(outputs, dim=1).squeeze(0).numpy()
    
    # Get corresponding true f(x) slice
    true_y = sample_y[:, :partial_length, :].squeeze(0).numpy()  # [partial_length]
    
    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(range(partial_length), partial_x.squeeze(), 'bo-', label="Partial g(x) Input")
    plt.plot(range(partial_length), true_y, 'go-', label="True f(x)")
    plt.plot(range(partial_length), pred_y, 'rx--', label="Predicted f(x)")
    plt.axvspan(0, partial_length-1, alpha=0.1, color='gray', label='Input Window')
    plt.legend()
    plt.title(f"Prediction from First {partial_length} Steps")
    plt.show()

test_partial_sequence(model, dataset, partial_length=5)
```

```python
def test_partial_sequence_at_index(model, dataset, start_idx=5, seq_len=5):
    """
    Predict f(x) for a slice of g(x) starting at any index.
    
    Args:
        model: Trained Seq2Seq model
        dataset: Dataset containing g(x) and f(x)
        start_idx: Starting index for the input sequence
        seq_len: Length of the sequence to predict
    """
    model.eval()
    
    # Get sample and add batch dimension
    sample_x, sample_y = dataset[0]  # sample_x: [full_seq_len, 1], sample_y: [full_seq_len, 1]
    sample_x = sample_x.unsqueeze(0)  # [1, full_seq_len, 1]
    sample_y = sample_y.unsqueeze(0)  # [1, full_seq_len, 1]

    # Validate indices
    assert start_idx >= 0, "Start index cannot be negative"
    assert start_idx + seq_len <= sample_x.shape[1], "Sequence extends beyond available data"
    
    # Take slice of g(x) starting at start_idx
    partial_x = sample_x[:, start_idx:start_idx+seq_len, :]  # [1, seq_len, 1]
    
    with torch.no_grad():
        # 1. Encode the partial sequence
        hidden, cell = model.encoder(partial_x)
        
        # 2. Initialize decoder
        decoder_input = torch.zeros((1, 1, 1), device=sample_x.device)
        outputs = []
        
        # 3. Decode autoregressively for the same length
        for t in range(seq_len):
            decoder_output, hidden, cell = model.decoder(decoder_input, hidden, cell)
            outputs.append(decoder_output)
            decoder_input = decoder_output  # Feed prediction back
        
        # Combine predictions: [1, seq_len, 1] -> [seq_len]
        pred_y = torch.cat(outputs, dim=1).squeeze(0).numpy()
    
    # Get corresponding true f(x) slice
    true_y = sample_y[:, start_idx:start_idx+seq_len, :].squeeze(0).numpy()  # [seq_len]
    
    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(range(start_idx, start_idx+seq_len), partial_x.squeeze(), 'bo-', 
             label=f"Input g(x) [{start_idx}:{start_idx+seq_len}]")
    plt.plot(range(start_idx, start_idx+seq_len), true_y, 'go-', 
             label="True f(x)")
    plt.plot(range(start_idx, start_idx+seq_len), pred_y, 'rx--', 
             label="Predicted f(x)")
    plt.axvspan(start_idx, start_idx+seq_len-1, alpha=0.1, color='gray', 
                label='Input Window')
    plt.legend()
    plt.title(f"Prediction from Index {start_idx} (Length={seq_len})")
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()
    
    return pred_y

test_partial_sequence_at_index(model, dataset, start_idx=10, seq_len=5)
```

```python
def test_partial_sequence(model, dataset, partial_length=5):
    model.eval()
    sample_x, sample_y = dataset[0]
    sample_x = sample_x.unsqueeze(0)  # Add batch dimension
    
    # Take only the first `partial_length` steps of g(x)
    partial_x = sample_x[:, :partial_length, :]
    
    with torch.no_grad():
        # Encode the partial sequence
        hidden, cell = model.encoder(partial_x)
        
        # Initialize decoder with zeros (or use a start token)
        decoder_input = torch.zeros((1, 1, 1), device=sample_x.device)
        outputs = []
        
        # Autoregressively generate the full sequence
        for t in range(len(sample_x.squeeze())):
            decoder_output, hidden, cell = model.decoder(decoder_input, hidden, cell)
            outputs.append(decoder_output)
            decoder_input = decoder_output  # Feed prediction back
        
        pred_y = torch.cat(outputs, dim=1).squeeze().numpy()
    
    # Plot results
    plt.figure(figsize=(10, 5))
    plt.plot(range(len(sample_x.squeeze())), sample_x.squeeze(), label="Full g(x) Input")
    plt.plot(range(len(sample_y.squeeze())), sample_y.squeeze(), label="True f(x) Output")
    plt.plot(range(len(pred_y)), pred_y, label="Predicted f(x)", linestyle='dashed')
    
    # Highlight the partial input used
    plt.axvspan(0, partial_length-1, alpha=0.2, color='gray', label=f'Partial g(x) (first {partial_length} steps)')
    plt.legend()
    plt.title(f"Prediction from Partial Input (First {partial_length} Steps)")
    plt.show()

# Test with first 5 timesteps
test_partial_sequence(model, dataset, partial_length=5)
```

```python
# Evaluation on full sequences
model.eval()
test_dataset = ChunkedFunctionDataset(num_samples=10, full_seq_length=100, chunk_size=20)

def predict_full_sequence(model, dataset, sequence_idx=0):
    """Predict full sequence by processing chunks and combining results"""
    g_full, f_full = dataset.get_full_sequence(sequence_idx)
    g_full = g_full.unsqueeze(0)  # Add batch dimension
    
    # Process in chunks
    predicted_chunks = []
    for i in range(0, len(g_full.squeeze()) - dataset.chunk_size + 1, dataset.step_size):
        chunk = g_full[:, i:i+dataset.chunk_size, :]
        with torch.no_grad():
            pred_chunk = model(chunk).squeeze().numpy()
        predicted_chunks.append((i, pred_chunk))
    
    # Combine predictions with averaging
    full_prediction = np.zeros(len(g_full.squeeze()))
    counts = np.zeros(len(g_full.squeeze()))
    for start, chunk in predicted_chunks:
        end = start + len(chunk)
        full_prediction[start:end] += chunk
        counts[start:end] += 1
    full_prediction /= counts
    
    # Plot results
    plt.figure(figsize=(12, 6))
    plt.plot(g_full.squeeze(), label="g(x) Input")
    plt.plot(f_full.squeeze(), label="True f(x)")
    plt.plot(full_prediction, label="Predicted f(x)", linestyle='dashed')
    plt.legend()
    plt.title("Full Sequence Prediction from Chunks")
    plt.show()

# Test prediction
predict_full_sequence(model, test_dataset, sequence_idx=0)
```
